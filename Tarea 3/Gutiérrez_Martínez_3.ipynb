### <center>Encontrar los puntos estacionarios de las siguientes funciones</center>

$$1)\; y=x^{3}-4x^{2}+x+6$$

1. Derivar función <br>
    $\frac{dy}{dx} = 3x^{2}-8x+1$<br><br>
    
2. Igualar derivada a $0$ y Obtener raíces <br>
    $3x^{2}-8x+1=0$

Obtener raíces<br>
$x_{1}=\frac{-b+\sqrt{b^{2}-4ac}}{2a}$<br>
$x_{1}=\frac{8+\sqrt{8^{2}-4(3)(1)}}{2(3)}$<br>
$x_{1}=2.5351$<br><br>

$x_{2}=\frac{-b-\sqrt{b^{2}-4ac}}{2a}$<br>
$x_{2}=\frac{8-\sqrt{64-12}}{6}$<br>
$x_{2}=0.1314$<br><br>

3. Sustituir las raíces en la ecuación original<br>
$y_{1}=2.5351^{3}-4(2.5351)^{2}+2.5351+6=-0.8794$<br>
$y_{2}=0.1314^{3}-4(0.1314)^{2}+0.1314+6=6.0646$<br><br>

por lo tanto hay puntos estacionarios en $(2.5351, -0.8794)$ y $(0.1314, 6.0646)$<br><br>

$$2)\; y=6x^{3}-12x^{2}+5x-2$$

1. Derivar función <br>
    $\frac{dy}{dx} = 18x^{2}-24x+5$<br><br>
    
2. Igualar derivada a $0$ y Obtener raíces <br>
    $18x^{2}-24x+5 = 0$

Obtener raíces<br>
$x_{1}=\frac{-b+\sqrt{b^{2}-4ac}}{2a}$<br>
$x_{1}=\frac{24+\sqrt{(-24)^{2}-4(18)(5)}}{2(18)}$<br>
$x_{1}=1.07491$<br><br>
$x_{2}=\frac{-b-\sqrt{b^{2}-4ac}}{2a}$<br>
$x_{2}=\frac{24-\sqrt{576-360}}{36}$<br>
$x_{2}=0.25841$<br><br>

3. Sustituir las raíces en la ecuación original<br>
$y_{1}=6(1.07491)^{3}-12(1.07491)^{2}+5(1.07491)-2 = -3.03871$<br>
$y_{2}=6(0.25841)^{3}-12(0.25841)^{2}+5(0.25841)-2 = -1.40572$<br><br>

por lo tanto hay puntos estacionarios en $(1.07491, -3.03871)$ y $(0.25841, -1.40572)$<br><br>

$$3)\; y=(1/3)x^{3}+8x^{2}+63x+7$$

1. Derivar función <br>
    $\frac{dy}{dx} = x^{2}+16x+63$<br><br>
    
2. Igualar derivada a $0$ y Obtener raíces <br>
    $x^{2}+16x+63 = 0$

Obtener raíces<br>
$x_{1}=\frac{-b+\sqrt{b^{2}-4ac}}{2a}$<br>
$x_{1}=\frac{-16+\sqrt{(16)^{2}-4(1)(63)}}{2(1)}$<br>
$x_{1}=-7$<br><br>

$x_{2}=\frac{-b-\sqrt{b^{2}-4ac}}{2a}$<br>
$x_{2}=\frac{-16-\sqrt{256-252}}{2}$<br>
$x_{2}=-9$<br><br>

3. Sustituir las raíces en la ecuación original<br>
$y_{1}=(1/3)(-7)^{3}+8(-7)^{2}+63(-7)+7 = -156.33$<br>
$y_{2}=(1/3)(-9)^{3}+8(-9)^{2}+63(-9)+7 = -155$<br><br>

por lo tanto hay puntos estacionarios en $(-7, -156.33)$ y $(-9, -155)$<br><br>
